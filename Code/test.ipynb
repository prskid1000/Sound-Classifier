```python
from keras.models import Sequential
from keras.layers import Conv2D, MaxPooling2D
from keras.layers import Activation, Dropout, Flatten, Dense
from sklearn.preprocessing import LabelEncoder
from keras.utils import np_utils as ut
import numpy as np
import matplotlib.pyplot as plt
import os
from keras import backend as K
import  keras.callbacks as cb

import tensorflow as tf
from tensorflow.python.tools import freeze_graph, optimize_for_inference_lib

tf.compat.v1.disable_eager_execution()

p=""
p=input("Enter the path where you extracted aimage folder ending with / = ")
model = Sequential()
model.add(Conv2D(64, (4, 4), input_shape=(64,64,4)))
model.add(Activation('relu'))
model.add(MaxPooling2D(pool_size=(2, 2)))

model.add(Conv2D(32, (4, 4)))
model.add(Activation('relu'))
model.add(MaxPooling2D(pool_size=(2, 2)))

model.add(Conv2D(8, (4, 4)))
model.add(Activation('relu'))
model.add(MaxPooling2D(pool_size=(2, 2)))


model.add(Flatten())
model.add(Activation('relu'))
model.add(Dense(1024))
model.add(Activation('relu'))
model.add(Dense(50))
model.add(Activation('relu'))

model.compile(loss='mean_squared_error',
              optimizer='sgd',
              metrics=['mae', 'acc'])
              
model_json=model.to_json()
with open(p+'aimage/model.json','w') as json_file:
 json_file.write(model_json)
 
x=np.load(p+'val_data.npz')
label=LabelEncoder()
labels=label.fit_transform(np.array(x['y']))
b=ut.to_categorical(labels)

model.fit(x['x'],b,epochs=10000,callbacks = [cb.TensorBoard(log_dir=p)], validation_split = 0.2)
model.save_weights(p+"aimage/data_weights.h5")

print(model.inputs)
print(model.outputs)
model.summary()
```

```python
from keras.models import Sequential
from keras.layers import Conv2D, MaxPooling2D
from keras.layers import Activation, Dropout, Flatten, Dense
from sklearn.preprocessing import LabelEncoder
from keras.utils import np_utils as ut
import numpy as np
import matplotlib.pyplot as plt
import os
from keras import backend as K
import  keras.callbacks as cb
import tensorflow as tf
from keras.models import model_from_json

p=""
p=input("Enter the path where you extracted aimage folder ending with / = ")

json_file = open(p+'aimage/model.json', 'r')
loaded_model_json = json_file.read()
json_file.close()
model = model_from_json(loaded_model_json)
model.load_weights(p+'aimage/data_weights.h5')

model.compile(loss='mean_squared_error',
              optimizer='sgd',
              metrics=['mae', 'acc'])


x=np.load(p+'val_data.npz')
label=LabelEncoder()
labels=label.fit_transform(np.array(x['y']))
b=ut.to_categorical(labels)

model.fit(x['x'],b,epochs=1, validation_split = 0.2)
              
model_json=model.to_json()
with open(p+'aimage/model.json','w') as json_file:
 json_file.write(model_json)
model.save_weights(p+"aimage/data_weights.h5")

print(model.inputs)
print(model.outputs)
model.summary()
```
